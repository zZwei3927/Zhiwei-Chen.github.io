```python
import unicodecsv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb

# 读取文件
df_gz2017 = pd.read_csv(r"C:\Users\WGW\Desktop\Jupyter Notebook\Data\Guangzhou_2017_PM2.5.csv")
# 重新定义列标签
df_gz2017.columns = ['Site', 'Parameter', 'Date', 'Year', 'Month', 'Day', 'Hour', 'Value', 'QC']

# 显示前几条数据
df_gz2017.head()

# 数据简要描述
df_gz2017.describe()
```

```python
# 查看缺失数据的总数目
len(df_gz2017.ix[df_gz2017.Value<0, :])   # iloc通过行号获取行数据，ix通可过行号&行标签获取行数据
```

```python
# 把错误设置为空值
df_gz2017.loc[df_gz2017.Value<0, 'Value'] = np.nan   # loc检索行标签,添加列指标

# 删除空值
df_gz2017.dropna(inplace=True)

# 数据简要描述
df_gz2017.describe()
```

```python
# 添加标签（空气等级）的定义
def get_grade(value):
    if value <= 50 and value >= 0:
        return 'Good'
    elif value <= 100:
        return 'Moderate'
    elif value <= 150:
        return 'Unhealthy for Sensi'
    elif value <= 200:
        return 'Unhealthy'
    elif value <= 300:
        return 'Very Unhealthy'
    elif value <= 500:
        return 'Hazardous'
    elif value > 500:
        return 'Beyond Index'   # 爆表了
    else:
        return None   # 输入值无效
    
# 添加列标签，并赋值
df_gz2017.loc[:, 'Grade'] = df_gz2017['Value'].apply(get_grade)

# 显示前五条数据
df_gz2017.head()
```

```python
# 添加2016成都数据
df_cd2016 = pd.read_csv('r"C:\Users\WGW\Desktop\Jupyter Notebook\Data\Chengdu_2016_PM2.5.csv')

df_cd2016.head()

# 重新定义列标签
df_cd2016.columns = ['Site', 'Parameter', 'Date', 'Year', 'Month', 'Day', 'Hour', 'Value', 'QC']

df_cd2016.describe()

# 统计错误数据
len(df_cd2016.ix[df_cd2016.Value<0, :])

# 把错误数据设置为空值
df_cd2016.loc[df_cd2016.Value<0, 'Value'] = np.nan

# 删除空值
df_cd2016.dropna(inplace=True)

df_cd2016.head()

df_cd2016.describe()

# 添加标签的定义
def get_grade(value):
    if value <= 50 and value >= 0:
        return 'Good'
    elif value <= 100:
        return 'Moderate'
    elif value <= 150:
        return 'Unhealthy for Sensi'
    elif value <= 200:
        return 'Unhealthy'
    elif value <= 300:
        return 'Very Unhealthy'
    elif value <= 500:
        return 'Hazardous'
    elif value > 500:
        return 'Beyond Index'   # 爆表了
    else:
        return None   # 输入值无效
    
# 添加列标签，并赋值
df_cd2016.loc[:, 'Grade'] = df_cd2016['Value'].apply(get_grade)

df_cd2016.head()
```

```python
# 添加2016广州数据
df_gz2016 = pd.read_csv('r"C:\Users\WGW\Desktop\Jupyter Notebook\Data\Guangzhou_2016_PM2.5.csv')

df_gz2016.head()

# 重新定义列名
df_gz2016.columns = ['Site', 'Parameter', 'Date', 'Year', 'Month', 'Day', 'Hour', 'Value', 'QC']

df_gz2016.describe()

# 统计错误数据
len(df_gz2016.ix[df_gz2016.Value<0, :])

# 把错误数据设置为空值
df_gz2016.loc[df_gz2016.Value<0, 'Value'] = np.nan

#删除错误数据
df_gz2016.dropna(inplace=True)

df_gz2016.head()

df_gz2016.describe()

#添加标签的定义
def get_grade(value):
    if value <= 50 and value >= 0:
        return 'Good'
    elif value <= 100:
        return 'Moderate'
    elif value <= 150:
        return 'Unhealthy for Sensi'
    elif value <= 200:
        return 'Unhealthy'
    elif value <= 300:
        return 'Very Unhealthy'
    elif value <= 500:
        return 'Hazardous'
    elif value > 500:
        return 'Beyond Index'   # 爆表了
    else:
        return None   # 输入值无效
    
# 添加列标签，并赋值
df_gz2016.loc[:, 'Grade'] = df_gz2016['Value'].apply(get_grade)

df_gz2016.head()
```

```python
# 统计全年各个等级的次数占比
gz2016_grade = df_gz2016.groupby(['Grade']).size() / len(df_gz2016)   # size()计算数组和矩阵所有数据的个数
gz2017_grade = df_gz2017.groupby(['Grade']).size() / len(df_gz2017)
cd2016_grade = df_cd2016.groupby(['Grade']).size() / len(df_cd2016)
cd2017_grade = df_cd2017.groupby(['Grade']).size() / len(df_cd2017)

# 定义一个空气质量等级索引
ix_grade = ['Good', 'Moderate', 'Unhealthy for Sensi', 'Unhealthy', 'Very Unhealthy', 'Hazardous', 'Beyond Index']

# 创建一个DataFrame对象，先把广州2016年的空气质量等级占比数据加进去
df_grade = pd.DataFrame(gz2016_grade, index=ix_grade, columns=['gz2016'])

df_grade['gz2017'] = gz2017_grade
df_grade['cd2016'] = cd2016_grade
df_grade['cd2017'] = cd2017_grade

# 查看我们获得的DataFrame，这里包含了两地两年的空气质量等级占比数据
df_grade
```

```python
# 使用饼图查看广州空气质量等级占比
# pie()饼图
df_grade.ix[:, 'gz2016'].plot.pie(title='Guangzhou 2016 AQI', autopct='%.1f%%', fontsize=12, figsize=(6, 6))

df_grade.ix[:, 'gz2017'].plot.pie(title='Guangzhou 2017 AQI', autopct='%.1f%%', fontsize=12, figsize=(6, 6))

# 使用柱状图查看空气质量等级对比
# bar()柱状图
df_grade.ix[:, ['gz2016', 'gz2017']].plot.bar(title='Guangzhou AQI 2016 vs 2017', figsize=(8, 6), fontsize=12)

df_grade.ix[:, ['gz2017', 'cd2017']].plot.bar(title='2017 AQI Guangzhou vs Chengdu', figsize=(8, 6), fontsize=12)

df_grade.ix[:, ['cd2016', 'cd2017']].plot.bar(title='Chengdu AQI 2016 vs 2017', figsize=(8, 6), fontsize=12)
```

```python
# 月度趋势

# 计算两地两年的pm2.5测量值月度平均值（mean()）
gz2017_month = df_gz2017.groupby(['Month'])['Value'].mean()
gz2016_month = df_gz2016.groupby(['Month'])['Value'].mean()
cd2017_month = df_cd2017.groupby(['Month'])['Value'].mean()
cd2016_month = df_cd2016.groupby(['Month'])['Value'].mean()

# 把平均值放到同一个DataFrame中
df_month = pd.DataFrame({'gz2016':gz2016_month}, index=np.arange(1, 13))

df_month['gz2017'] = gz2017_month
df_month['cd2016'] = cd2016_month
df_month['cd2017'] = cd2017_month

df_month
```

```python
# 对比广州2016和2017的月度数据
# plot()折线图
df_month.ix[:, ['gz2016', 'gz2017']].plot(title='Guangzhou PM2.5 Monthly Avg. 2016 vs 2017', figsize=(8, 4))

df_month.ix[:, ['cd2016', 'cd2017']].plot(title='Chengdu PM2.5 Monthly Avg. 2016 vs 2017', figsize=(8, 4))

df_month.ix[:, ['gz2016', 'cd2016']].plot(title='2016 PM2.5 Monthly Avg. Chengdu vs Guangzhou', figsize=(8, 4))

df_month.ix[:, ['gz2017', 'cd2017']].plot(title='2017 PM2.5 Monthly Avg. Chengdu vs Guangzhou', figsize=(8, 4))
```

```python
# 每小时数据分析
df_hour = pd.DataFrame({'Month':df_gz2016.ix[:, 'Month'], 
                        'Day':df_gz2016.ix[:, 'Day'], 
                        'Hour':df_gz2016.ix[:, 'Hour'], 
                        'gz2016':df_gz2016.ix[:, 'Value']})
df_hour.describe()
df_hour.head()

# 把广州2017数据添加到数据框
# merge(,on:用于连接的列名)合并数据集
df_hour = df_hour.merge(df_gz2017.ix[:, ['Month', 'Day', 'Hour', 'Value']], on=('Month', 'Day', 'Hour'))

# 对于value列重命名
df_hour.rename_axis({'Value':'gz2017'}, axis="columns", inplace=True)
df_hour.describe()

# 把成都2016数据添加到数据框
df_hour = df_hour.merge(df_cd2016.ix[:, ['Month', 'Day', 'Hour', 'Value']], on=('Month', 'Day', 'Hour'))
df_hour.rename_axis({'Value':'cd2016'}, axis="columns", inplace=True)

# 把成都2017数据添加到数据框
df_hour = df_hour.merge(df_cd2017.ix[:, ['Month', 'Day', 'Hour', 'Value']], on=('Month', 'Day', 'Hour'))
df_hour.rename_axis({'Value':'cd2017'}, axis="columns", inplace=True)

df_hour.describe()
df_hour.head()

# 对比广州2016和2017的数据
len(df_hour[df_hour['gz2016']>df_hour['gz2017']]), 1.0*len(df_hour[df_hour['gz2016']>df_hour['gz2017']])/len(df_hour)
len(df_hour[df_hour['gz2016']<df_hour['gz2017']]), 1.0*len(df_hour[df_hour['gz2016']<df_hour['gz2017']])/len(df_hour)

# 对比成都2016和广州2016的数据
len(df_hour[df_hour['cd2016']>df_hour['gz2016']]), 1.0*len(df_hour[df_hour['cd2016']>df_hour['gz2016']])/len(df_hour)
len(df_hour[df_hour['cd2016']<df_hour['gz2016']]), 1.0*len(df_hour[df_hour['cd2016']<df_hour['gz2016']])/len(df_hour)
```

```python
# 画图对比广州2016和2017数据
# plot()折线图
df_hour.ix[:, ['gz2016', 'gz2017']].plot(title='Guangzhou PM2.5 Hourly 2016 vs 2017', figsize=(12, 4))

df_hour.ix[:, ['cd2016', 'cd2017']].plot(title='Chengdu PM2.5 Hourly 2016 vs 2017', figsize=(12, 4))

df_hour.ix[:, ['cd2016', 'gz2016']].plot(title='2016 PM2.5 Hourly Chengdu vs Guangzhou', figsize=(12, 4))

df_hour.ix[:, ['cd2017', 'gz2017']].plot(title='2017 PM2.5 Hourly Chengdu vs Guangzhou', figsize=(12, 4))
```
